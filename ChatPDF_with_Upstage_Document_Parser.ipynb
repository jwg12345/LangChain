```python
import gradio as gr
from gradio.themes.base import Base
from gradio_pdf import PDF

import os
from langchain_upstage import UpstageDocumentParseLoader
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory


class Seafoam(Base):
    pass
seafoam = Seafoam()

chat_history_for_chain = ChatMessageHistory()
chat = ChatOpenAI(model="gpt-4o")

with gr.Blocks(theme=seafoam) as demo: #'JohnSmith9982/small_and_pretty'
    def question_generator(pdf, chatbot_history):
        loader = UpstageDocumentParseLoader(pdf, split='page')

        pages = loader.load()  # or loader.lazy_load()
        pdf_context = ''
        for page in pages:
            pdf_context += page.page_content

        prompt_text = f"""
            You are an advanced question generator designed to create a meaningful and contextually relevant question from provided text.
            Use the given text to generate a thought-provoking or comprehension-based question IN KOREAN.
            Do not include answers.
            {pdf_context}
        """
        chat_history_for_chain.add_ai_message(prompt_text)

        chatbot_history.append(("", chat.invoke(prompt_text).content))

        return chatbot_history


    def response(message, chatbot_history):
        prompt = ChatPromptTemplate.from_messages(
            [
                (
                    "system",
                    """
                    If user answered correctly, please give the next question.
                    If user did not answer or got it wrong, kindly explain it based on the text provided.
                    YOU MUST ANSWER IN KOREAN.""",
                ),
                ("placeholder", "{chat_history}"),
                ("human", "{input}"),
            ]
        )
        chain = prompt | chat

        chain_with_message_history = RunnableWithMessageHistory(
            chain, # 실행할 Runnable 객체
            lambda session_id: chat_history_for_chain, # 세션 기록을 가져오는 함수
            input_messages_key="input", # 입력 메시지의 Key
            history_messages_key="chat_history", # 대화 히스토리 메시지의 Key
        )

        response = chain_with_message_history.invoke(
            {"input": message},
            {"configurable": {"session_id": "123"}},
        )

        chatbot_history.append((message, response.content))
        return chatbot_history

    with gr.Row():
        with gr.Column(scale=4):
            gr.HTML("""<div style="text-align: center; max-width: 1000px; margin: 10px auto;">
                <div>
                    <h1>📖 PDF로 AI와 공부하기 📖</h1>
                </div>
                <p style="margin-bottom: 10px; font-size: 95%">
                    💭 학습자료 PDF를 업로드 해보세요. AI가 자료에서 질문을 만들어 줄거예요. 대화를 주고 받아 보세요.</a>
                </p>
            </div>""")


    with gr.Row():
        with gr.Column(scale = 1):
            pdf = PDF(label="Upload a PDF", interactive=True)
            name = gr.Textbox(label="PDF 파일명")
            pdf.upload(lambda f: f, pdf, name)
            btn = gr.Button("PDF 제출", variant="primary")

        with gr.Column(scale = 3):
            chatbot = gr.Chatbot()

            btn.click(question_generator, [pdf, chatbot], [chatbot])

            msg = gr.Textbox(placeholder="AI가 출제한 문제에 답을 하거나 질문을 해보세요.", label="입력")

            with gr.Row():
                gr.HTML("""<div style="text-align: center; max-width: 500px; margin: 0px auto;">
                    <div>
                        <h1>  </h1>
                    </p>
                </div>""")
                gr.HTML("""<div style="text-align: center; max-width: 500px; margin: 0px auto;">
                    <div>
                        <h1>  </h1>
                    </p>
                </div>""")
                btn = gr.Button("Submit", variant="primary")
                clear = gr.Button("Clear")

    btn.click(fn=response, inputs=[msg, chatbot], outputs=[chatbot])
    msg.submit(response, [msg, chatbot], [chatbot])

    clear.click(lambda: None, None, msg, queue=False)

demo.launch(debug=True, share=True)#debug=True, share=True
```
